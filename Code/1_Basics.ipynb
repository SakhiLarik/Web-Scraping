## Load our first page

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup as bs
```

```python
# Load the page content
url = "http://localhost:5500/Pages/Rapid"
req = requests.get(url)
req
```

```python
if(req.status_code == 200):
    print("Succesfully loaded the page")
```

```python
# Convert into Beautiful Soap
soup = bs(req.content)
print(soup)
```

## Find and Find_all

```python
# Getting Title
title = soup.find_all('title')
print(title[0].text)
```

```python
# Geting all headings
headings_1 = []
for heading in soup.find_all("h1"):
    headings_1.append(heading)
print(headings_1)
```

```python
# Geting links of headings
heading_links = [links.a for links in headings_1]
heading_links
```

```python
# Geting text of headings
heading_text = [text.text for text in headings_1]
heading_text
```

### Get List of elements

```python
all_headings = soup.find_all(["h1","h2","h3","h4","h5","h6"])
all_headings
```

```python
# Get Headings that has class as title
title_headings = soup.find_all(["h4","h2","h3","h1","h5","h6"],attrs={"class":"title"})
print(title_headings)
```

```python
titles = [title.text for title in title_headings]
titles
```

### Nesting Finder

```python
# Get span text from headings
span = soup.find("h2").find("span")
print(span.text)
```

```python
h1 = soup.find("h1")
a = h1.find("a")
span = a.find("span")
print(span.text)
#OR
```

```python
soup.find("h1").find("a").find("span").text
```

```python
# Get div that has heading in it
soup.find("body").find(["h1","h2","h3","h4","h5","h6"]).find("a")
```

### Find the parent of a Tag

```python
serviceHeading = soup.find_all(["h1","h2","h3","h4","h5","h6"],string="Services")
section_header = serviceHeading[0].parent
container = section_header.parent
print(container.prettify())
```

### Collecting servics and creating a PD DF from it and display the data

```python
titles = [title.text for title in soup.find_all("h4",attrs={"class":"title"})]
titles
```

```python
description = [title.text for title in soup.find_all("p",attrs={"class":"description"})]
description
```

```python
df = pd.DataFrame({"Titles":titles,"Description":description})
df.astype("str")
df.head()
```

# SELECT
The select works same as find_all but it takes arguments as in CSS we do for styling

```python
all_div = soup.select("div")
all_div
```

```python
# Get all the icon tags with class of .fa
icons = soup.select("i.fa")
icons
```

```python
# Get All description tags
desc = soup.select(".description")
desc
```

```python
# Get all heading tags
h = soup.select("h1, h2, h3, h4, h5, h6")
h
```

```python
# Get images that has class img-fluid
imgs = soup.select("img.img-fluid")
imgs
```

## Get a specific property of an element

```python
# Get all the links from anchor tags
a = soup.select("a")
links = [link['href'] for link in a if (link['href'] != '#' and link['href'] != '' and not link['href'].startswith("#"))]
links
```

##### Other Programmers and Developers

```python
# Get all the suitable links from anchor tags
a = soup.select("a")
links = []
for link in a:
    link_href = link['href']
    if(link_href != "#"):
        if(link_href != ""):
            if(not link_href.startswith("#")):
                links.append(link_href)
print(links)
```
