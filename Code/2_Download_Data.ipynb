```python
from bs4 import BeautifulSoup as bs
import requests
```

```python
url = "https://www.infogrepper.com/image-url-for-testing/"
req = requests.get(url)
req
```

```python
soup = bs(req.content)
soup
```

```python
imgs = soup.select("img")
imgs
```

```python
links = [url['src'] for url in imgs ]
links
```

```python
# I just want to save alternative images
img_no = 1
for link in links[::2]:
    img_content = requests.get(link).content
    with open(f"../img/{img_no}.png",'wb') as img:
        img.write(img_content)
    img_no = img_no + 1
```
